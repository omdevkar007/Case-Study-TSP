```python
import pandas as pd

# Load dataset
df = pd.read_csv("ZOMATO STATS.csv")

# Convert 'rate' column to numeric by extracting numeric values and converting to float
df["rate"] = df["rate"].str.extract(r"([\d.]+)").astype(float)

# Convert numeric columns to appropriate data types
numeric_cols = ["votes", "approx_cost(for two people)", "rate"]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")

# Handling missing values by filling with mean values
df.fillna(df.mean(numeric_only=True), inplace=True)

# Display dataset summary
df.info()
df.describe()
```

#**LINE PLOTTING**#

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(30, 15))
plt.plot(df['votes'], df['rate'], marker='d', linestyle='--', markersize=10, color='green', alpha=1)
plt.title('Restaurant Ratings Trend w.r.t. Votes', fontsize=30)
plt.xlabel('Votes', fontsize=30)
plt.ylabel('Ratings', fontsize=30)
plt.xticks(rotation=45, fontsize=20)
plt.yticks(rotation=45, fontsize=20)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(22, 11))
plt.plot(df['votes'], df['rate'], marker='D', linestyle=':', markersize=10, color='orange', alpha=1)
plt.title('Ratings Trend w.r.t. Votes', fontsize=30)
plt.xlabel('Votes', fontsize=30)
plt.ylabel('Ratings', fontsize=30)
plt.xticks(rotation=45, fontsize=20)
plt.yticks(rotation=45, fontsize=20)
plt.grid()
plt.show()

print()
plt.figure(figsize=(22, 11))
plt.plot(df['votes'], df['approx_cost(for two people)'], marker='p', linestyle='--', markersize=10, color='blue', alpha=1)
plt.title('Cost Trend w.r.t. Votes', fontsize=30)
plt.xlabel('Votes', fontsize=30)
plt.ylabel('Cost for Two People', fontsize=30)
plt.xticks(rotation=45, fontsize=20)
plt.yticks(rotation=45, fontsize=20)
plt.grid()
plt.show()

print()
plt.figure(figsize=(22, 11))
plt.plot(df['approx_cost(for two people)'], df['rate'], marker='h', linestyle='-.', markersize=10, color='darkorange', alpha=1)
plt.title('Ratings Trend w.r.t. Cost', fontsize=30)
plt.xlabel('Cost for Two People', fontsize=30)
plt.ylabel('Ratings', fontsize=30)
plt.xticks(rotation=45, fontsize=20)
plt.yticks(rotation=45, fontsize=20)
plt.grid()
plt.show()
```

**PIECHART**

```python
import matplotlib.pyplot as plt

# Get unique restaurant names
restaurants = df["name"].unique()
values = [df[df["name"] == restaurant]["rate"].mean() for restaurant in restaurants[:5]]  # Taking only the first 5 restaurants for better visualization

# Plot pie chart
plt.pie(values, labels=restaurants[:5], radius=1, autopct='%1.2f%%', shadow=True, explode=[0.1] * 5, startangle=90)
plt.title("Ratings Distribution Across 5 Restaurants")

plt.show()
```

**BOX PLOT**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Drop missing values for ratings
df = df.dropna(subset=["rate"])

# Box plot for Ratings
plt.figure(figsize=(1, 5))
sns.boxplot(y="rate", data=df)
plt.xlabel("Ratings")
plt.ylabel("Values")
plt.title("Box Plot of Ratings")
plt.show()
```

**JOINT PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot for Votes vs Ratings
plot1 = sns.jointplot(x="votes", y="rate", data=df, kind="scatter", color="blue")
plt.show()

# Regression plot for Votes vs Ratings
plot2 = sns.jointplot(x="votes", y="rate", data=df, kind="reg", color="red")
plt.show()

# KDE plot for Votes vs Ratings
plot3 = sns.jointplot(x="votes", y="rate", data=df, kind="kde", cmap="Purples")
plt.show()
```

**COUNT PLOT**

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.countplot(x="name", data=df, palette="magma", order=df["name"].value_counts().index[:10])  # Top 10 restaurants
plt.xticks(rotation=90)
plt.xlabel("Restaurant Name")
plt.ylabel("Count")
plt.title("Count Plot: Number of Data Points per Restaurant")
plt.show()
```
